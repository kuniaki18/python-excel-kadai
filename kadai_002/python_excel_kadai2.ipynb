```python
import pandas as pd

# Step1. データフレームを作成
data = {
    "日付": ["2023-05-17", "2023-05-18", "2023-05-19", "2023-05-20", "2023-05-21"],
    "社員名": ["山田", "佐藤", "鈴木", "田中", "高橋"],
    "売上": [100, 200, 150, 300, 250],
    "部門": ["メーカー", "代理店", "メーカー", "商社", "代理店"]
}

df = pd.DataFrame(data)

# Step2. 全体の平均売上を算出
overall_avg = df["売上"].mean()

# Step3. 「平均売上」列を作成（全員同じ値）
df["平均売上"] = overall_avg

# Step4. 業績ランクを付与する関数
def performance(sales, avg):
    if sales >= avg + 50:
        return "A"
    elif sales >= avg:
        return "B"
    else:
        return "C"

# Step5. 業績ランク列を作成
df["業績ランク"] = df["売上"].apply(lambda x: performance(x, overall_avg))

# Step6. Excel に書き出し
df.to_excel("業績.xlsx", index=False)

print("業績.xlsx を作成しました。")
```
