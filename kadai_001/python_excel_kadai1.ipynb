```python
!pip install openpyxl

from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, Border, Side
from datetime import datetime

# ============================
# ① 新規Excelファイル作成
# ============================
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# ============================
# ② 日付（現在日付）
# ============================
today = datetime.now().strftime("%Y/%m/%d")

# ============================
# ③ 固定データ
# ============================
ws["B2"] = "請求書"

ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678   FAX:03-1234-5678"
ws["B7"] = "担当者名: 鈴木一郎 様"
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today


# ============================
# ⑤ 表ヘッダー
# ============================
ws.append([""] * 6)  # 空行
headers = ["商品名", "数量", "単価", "金額"]
header_row = 10

for col, h in enumerate(headers, start=2):
    c = ws.cell(row=header_row, column=col, value=h)
    c.font = Font(bold=True)
    c.alignment = Alignment(horizontal="center")

# ============================
# ⑥ 商品データ（例）
# ============================
items = [
    ["商品A", 2, 10000],
    ["商品B", 1, 15000],
]

start_row = header_row + 1
row = start_row
subtotal = 0

for name, qty, price in items:
    amount = qty * price
    subtotal += amount

    ws.cell(row=row, column=2, value=name)
    ws.cell(row=row, column=3, value=qty)
    ws.cell(row=row, column=4, value=price)
    ws.cell(row=row, column=5, value=amount)


    row += 1

# ============================
# ⑦ 合計・消費税・税込合計（数値で書き込む）
# ============================
total_row = row + 1
tax_row = row + 2
grand_row = row + 3

tax = int(subtotal * 0.1)
grand_total = subtotal + tax

ws[f"B{total_row}"] = "合計"
ws[f"E{total_row}"] = subtotal

ws[f"B{tax_row}"] = "消費税"
ws[f"E{tax_row}"] = tax

ws[f"B{grand_row}"] = "税込合計"
ws[f"E{grand_row}"] = grand_total


# ============================
# ⑨ 保存
# ============================
filename = f"請求書_{datetime.now().strftime('%Y%m%d')}.xlsx"
wb.save(filename)

print("作成しました：", filename)
```
